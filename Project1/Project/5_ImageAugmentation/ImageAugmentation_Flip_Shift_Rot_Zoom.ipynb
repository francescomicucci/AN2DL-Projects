### Load and process the dataset

```python
# Load data
data = np.load('../clean_data.npz', allow_pickle=True)

# Save images in a new list (image pixel values are float in [0, 1])
images = data['data']

# Save labels in a new list
labels = data['labels']

# Print data shape
print(images.shape, labels.shape)
```

```python
# Convert labels from string to integer
num_labels = []
for label in labels:
  if label == "healthy":
    num_labels.append(0)
  else:
    num_labels.append(1)

# Convert labels in the one-hot encoding format
num_labels = np.array(num_labels)
labels = tfk.utils.to_categorical(num_labels, 2)

# Split data into train and val sets
X_train, X_val, y_train, y_val = train_test_split(
    images,
    labels,
    test_size=0.1,
    stratify=labels
)

# Print shapes of the datasets
print(f"X_train shape: {X_train.shape}, y_train shape: {y_train.shape}")
print(f"X_val shape: {X_val.shape}, y_val shape: {y_val.shape}")
```

### ConvNeXtLarge

```python
# Import Feature Extractor with specified settings
feature_extractor = tfk.applications.ConvNeXtLarge(
    input_shape=(96, 96, 3),
    include_top=False,
    weights="imagenet",
    pooling='avg',
    include_preprocessing=True,
)
```

### Model

```python
# Define network parameters/callbacks
dropout_rate = 1/10
learning_rate = 1e-3

lr_scheduler = tfk.callbacks.ReduceLROnPlateau(
    monitor='val_accuracy',
    patience=5,
    factor=0.999,
    mode='max',
    min_lr=1e-5
)

early_stopping = tfk.callbacks.EarlyStopping(
    monitor='val_accuracy',
    mode='max',
    patience=30,
    restore_best_weights=True
)
```

```python
def build_model(feature_extractor, dropout_rate, learning_rate=0.001):
  # Use the supernet as feature extractor, i.e. freeze all its weigths
  feature_extractor.trainable = False

  # Create an input layer with shape (96, 96, 3)
  inputs = tfk.Input(shape=(96, 96, 3), name='Input')

  # Connect FeatureExtractor to the input
  x = feature_extractor(inputs)

  # Hidden layers
  x = tfkl.Dense(units=256, name='HiddenDense1')(x)
  x = tfkl.Activation('relu', name='HiddenActivation1')(x)
  dropout = tfkl.Dropout(dropout_rate)(x)
  x = tfkl.Dense(units=256, name='HiddenDense2')(dropout)
  x = tfkl.Activation('relu', name='HiddenActivation2')(x)
  dropout = tfkl.Dropout(dropout_rate)(x)

  # Add a Dense layer with 2 unit and softmax activation as the classifier
  outputs = tfkl.Dense(2, activation='softmax')(dropout)

  # Create a Model connecting input and output
  model = tfk.Model(inputs=inputs, outputs=outputs, name='model')

  # Compile the model with Categorical Cross-Entropy loss and Adam optimizer
  model.compile(loss=tfk.losses.CategoricalCrossentropy(),
                optimizer=tfk.optimizers.Adam(learning_rate),
                metrics=['accuracy'])

  # Return the model
  return model
```

### FLIP + SHIFT + ROT + ZOOM

```python
model = build_model(feature_extractor, dropout_rate, learning_rate)
model.summary()
```

```python
train_datagen = ImageDataGenerator(
        horizontal_flip=True,
        vertical_flip = True,
        width_shift_range = 0.1,
        height_shift_range = 0.1,
        rotation_range=90.,
        zoom_range=0.1,
        fill_mode='reflect')
```

```python
history_flip_shift_rot_zoom = model.fit_generator(train_datagen.flow(X_train*255, y_train,
                              batch_size=16),
                              epochs=200,
                              steps_per_epoch=len(X_train)/16,
                              validation_data = (X_val*255, y_val),
                              callbacks = [early_stopping, lr_scheduler]
                              ).history
```

```python
# Plot the training histories
plt.figure(figsize=(15,5))
plt.plot(history_flip_shift_rot_zoom['loss'], alpha=.3, color='#4D61E2', linestyle='--')
plt.plot(history_flip_shift_rot_zoom['val_loss'], label='Filter', alpha=.8, color='#4D61E2')
plt.legend(loc='upper left')
plt.title('Categorical Crossentropy')
plt.grid(alpha=.3)

plt.figure(figsize=(15,5))
plt.plot(history_flip_shift_rot_zoom['accuracy'], alpha=.3, color='#4D61E2', linestyle='--')
plt.plot(history_flip_shift_rot_zoom['val_accuracy'], label='Filter', alpha=.8, color='#4D61E2')
plt.legend(loc='upper left')
plt.title('Accuracy')
plt.grid(alpha=.3)

plt.show()
```

```python
# Save the model
model.save('Augmentation_Flip_Shift_Rot_Zoom')
with open('Augmentation_Flip_Shift_Rot_Zoom/history.pkl', 'wb') as f:
  pickle.dump(history_flip_shift_rot_zoom, f)
```
